```python
def multivariate_gaussian(xx, mu, sigma2):
    m, n  = np.shape(sigma2)
    sigma2 = np.diag(sigma2[0, :])
    xx2 = xx - mu
    det = np.linalg.det(sigma2)
    inv = np.linalg.inv(sigma2)
    term_ = np.reshape((-0.5) * np.sum(np.multiply(xx2@inv, xx2), 1), (np.size(xx2, 0), 1))
    prob = np.power(2*pi, -n/2) * np.power(det, -0.5) * np.exp(term_)
    return prob
```

```python
def estimate_gaussian(xx):
    m,n = np.shape(xx)
    mu = 1/m * np.sum(xx,0)
    mu = np.reshape(mu, (1,n))
    sigma = 1/m * np.sum(np.power((xx-mu),2),0)
    sigma = np.reshape(sigma, (1,n))
    return mu, sigma
```

```python
data = np.zeros(shape=(len(tr.columns),2))

for i in range(0,len(tr.columns)):
    data[i,0] = tr.mean()[i]
    data[i,1] = tr.std()[i]

data = pd.DataFrame(data)
```

```python
data.columns = ['mean', 'std']
```

```python
x_val = data.values
```

```python
mu, sigma = estimate_gaussian(x_val)
```

```python
p = multivariate_gaussian(x_val, mu, sigma)
```
